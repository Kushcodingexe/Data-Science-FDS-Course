```python
from IPython.display import display
csv_path = "ds_salaries.csv"   # https://www.kaggle.com/datasets/ruchi798/data-science-job-salaries?resource=download
df = pd.read_csv(csv_path, low_memory=False)
print("Loaded:", csv_path, " — shape:", df.shape)
print()

```

9. Display the first 10 rows

```python
print("=== 9) First 10 rows ===")
display(df.head(10))
print()
```

10. Print names of all columns and their data types

```python
print("=== 10) Columns and dtypes ===")
print(df.dtypes)
print()
print("Column list:", list(df.columns))
print()
```

```python
# --- Helper: detect best salary column (prefer numeric USD column if present) ---
def detect_salary_column(df):
    candidates = [c for c in df.columns if any(k in c.lower() for k in ("salary","comp","convertedcomp","pay"))]
    # prefer salary_in_usd if present
    for pref in ("salary_in_usd", "salary_usd", "salary"):
        if pref in df.columns:
            return pref
    if candidates:
        return candidates[0]
    return None

salary_col = detect_salary_column(df)
print("Detected salary column:", salary_col)
print()
```

11. Extract only the salary column and calculate its mean (and median, count)

```python
print("=== 11) Salary statistics ===")
if salary_col is None:
    print("No salary-like column detected. Available columns:", list(df.columns))
else:
    # Clean and convert to numeric (remove commas, convert to floats)
    salary_series = pd.to_numeric(df[salary_col].astype(str).str.replace(',', ''), errors='coerce')
    mean_salary = salary_series.mean(skipna=True)
    median_salary = salary_series.median(skipna=True)
    count_nonnull = salary_series.notna().sum()
    print(f"Column used: {salary_col}")
    print("Count (non-null):", count_nonnull)
    print("Mean:", mean_salary)
    print("Median:", median_salary)
    # add cleaned numeric salary to dataframe for later use
    df['_salary_numeric'] = salary_series
print()
```

12. Display all rows by applying some filters on columns

```python
# Example filters: remote_ratio == 100 (fully remote) AND salary_in_usd above median
print("=== 12) Filtered rows example (fully remote & salary above median) ===")
# Choose realistic columns if present
filters = []
if 'remote_ratio' in df.columns:
    filters.append(df['remote_ratio'] == 100)
# use numeric salary if available
if '_salary_numeric' in df.columns:
    median_val = df['_salary_numeric'].median(skipna=True)
    filters.append(df['_salary_numeric'] > median_val)

if filters:
    combined = filters[0]
    for f in filters[1:]:
        combined &= f
    filtered_df = df.loc[combined].copy()
    print("Filtered shape:", filtered_df.shape)
    # show all rows of the filtered result — careful with very large results
    pd.set_option('display.max_rows', None)   # temporarily show all rows (use with caution)
    display(filtered_df)
    pd.reset_option('display.max_rows')
else:
    print("No appropriate columns found to demonstrate the filter. Available columns:", list(df.columns))
print()
```

13. Select rows from index 10 to 20 and display only specific columns

```python
print("=== 13) Rows index 10..20 with specific columns ===")
# Example columns (choose ones that typically exist)
example_cols = []
for c in ('work_year', 'experience_level', 'job_title', salary_col, 'salary_currency', 'employee_residence', 'company_size'):
    if c in df.columns or c == salary_col:
        example_cols.append(c)
# fallback to first 4 columns if none of the above are present
if not example_cols:
    example_cols = list(df.columns[:4])
subset_10_20 = df.iloc[10:21][example_cols]
display(subset_10_20)
print()
```

14. Replace values of a column when specific criteria are met

```python
# Example 1: Create/replace a 'pay_bracket' column marking top 10% as 'Top10%'
# Example 2: Replace short company_size codes with full labels (S->Small, M->Medium, L->Large)
print("=== 14) Replace values example ===")
if '_salary_numeric' in df.columns:
    thresh_90 = df['_salary_numeric'].quantile(0.90)
    df['pay_bracket'] = np.where(df['_salary_numeric'] >= thresh_90, 'Top10%', 'Normal')
    print("Added 'pay_bracket' based on top 10% threshold:", thresh_90)
    print(df['pay_bracket'].value_counts(dropna=False))
else:
    print("No numeric salary available; skipping pay_bracket example.")

# Replace company_size shorthand if present
if 'company_size' in df.columns:
    mapping = {'S': 'Small', 'M': 'Medium', 'L': 'Large'}
    # Use map with fillna to keep any unusual values unchanged
    df['company_size_full'] = df['company_size'].map(mapping).fillna(df['company_size'])
    print("Created 'company_size_full' with readable sizes. Sample:")
    display(df[['company_size', 'company_size_full']].head(10))
else:
    print("No 'company_size' column found; skipping that replacement example.")
print()
```

15. Sort the DataFrame by one column ascending and then by another column descending

```python
# Example: sort by 'employee_residence' (asc) then by salary numeric (desc)
print("=== 15) Sorting example ===")
sort_by = []
ascending = []
if 'employee_residence' in df.columns:
    sort_by.append('employee_residence'); ascending.append(True)
if '_salary_numeric' in df.columns:
    sort_by.append('_salary_numeric'); ascending.append(False)

if sort_by:
    sorted_df = df.sort_values(by=sort_by, ascending=ascending, na_position='last')
    print(f"Sorted by {sort_by} with ascending={ascending}. Showing top 20 rows:")
    display(sorted_df.head(20)[['employee_residence', '_salary_numeric', 'job_title', 'company_size_full'] if 'job_title' in df.columns else sorted_df.columns[:5]])
else:
    print("No suitable columns found to demonstrate sorting. Columns:", list(df.columns)[:10])

# --- Optional: Save modified DataFrame to a new CSV if you want to keep the changes ---
out_path = "ds_salaries_processed.csv"
df.to_csv(out_path, index=False)
print("\nSaved processed DataFrame to:", out_path)
```
